```
# Install required packages
!pip install librosa soundfile scipy noisereduce

import numpy as np
import librosa
import soundfile as sf
from scipy.signal import butter, filtfilt
import noisereduce as nr
from google.colab import files
from IPython.display import Audio, display, HTML
import tempfile
import os

def reduce_noise(y, sr, prop_decrease=0.85):
    """Reduce background noise using spectral gating."""
    return nr.reduce_noise(y=y, sr=sr, prop_decrease=prop_decrease)

def apply_lowpass_filter(y, sr, cutoff=5000):
    """Apply a low-pass filter to remove high-frequency artifacts."""
    nyquist = sr // 2
    normalized_cutoff = cutoff / nyquist
    b, a = butter(4, normalized_cutoff, btype='low')
    return filtfilt(b, a, y)

def process_audio(y, sr):
    """Process audio data with all enhancement techniques."""
    y_cleaned = reduce_noise(y, sr)
    y_filtered = apply_lowpass_filter(y_cleaned, sr)
    return librosa.util.normalize(y_filtered)

# Create a temporary directory for audio files
temp_dir = tempfile.mkdtemp()

print("Please upload an audio file (WAV, MP3, etc.)")
uploaded = files.upload()

for filename in uploaded.keys():
    # Load the original audio
    input_path = filename
    y, sr = librosa.load(input_path, sr=22050)

    # Process the audio
    y_processed = process_audio(y, sr)

    # Save processed audio to temp file
    output_path = os.path.join(temp_dir, 'processed_' + filename)
    sf.write(output_path, y_processed, sr)

    # Display audio players
    print("\nOriginal Audio:")
    display(Audio(y, rate=sr))

    print("\nProcessed Audio:")
    display(Audio(y_processed, rate=sr))

    # Add download button for processed audio
    with open(output_path, 'rb') as f:
        processed_audio_data = f.read()

    display(HTML(f"""
    <div style="margin: 20px 0;">
        <a href="data:audio/wav;base64,{processed_audio_data.hex()}"
           download="processed_{filename}"
           style="background-color: #4CAF50;
                  color: white;
                  padding: 10px 20px;
                  text-decoration: none;
                  border-radius: 5px;">
            Download Processed Audio
        </a>
    </div>
    """))

# Cleanup temporary directory
import shutil
shutil.rmtree(temp_dir)
```
